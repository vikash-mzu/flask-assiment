```python
Q1. What is Flask Framework? What are the advantages of Flask Framework?
Flask Framework: Flask is a lightweight and flexible web framework for Python. It is designed to be simple and easy to use, providing the necessary tools to build web applications without imposing a specific directory structure or dependencies.

Advantages of Flask:

Simplicity: Flask's minimalistic approach makes it easy to understand and use. It provides only the essential components, allowing developers to add more features as needed.
Flexibility: It offers flexibility in terms of project structure and extensions. Developers can choose how to structure their application and which libraries or tools to integrate.
Modularity: Flask supports modularity by allowing developers to build applications using blueprints and extensions, which helps in organizing code efficiently.
Lightweight: Being lightweight, Flask has a small footprint and does not come with unnecessary components. It provides just the basics to get started with web development.
Built-in Development Server: Flask includes a built-in development server, making it easy to test and debug applications locally.
Extensible: Flask has a wide range of extensions available for adding functionalities like authentication, database integration, form handling, and more.




Q2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in Jupyter Notebook.
Here is a simple Flask application code to display ‘Hello World!!’:


from flask import Flask

app = Flask(__name__)

@app.route('/')
def hello_world():
    return 'Hello World!!'

if __name__ == '__main__':
    app.run(debug=True)
Steps to Run the Application and Capture the Screenshot:

Save the code in a file named app.py.
Open a terminal or command prompt and navigate to the directory containing app.py.
Run the application using the command: python app.py.
Open a web browser and navigate to http://127.0.0.1:5000/.
Capture a screenshot of the output displaying Hello World!! and insert it into the Jupyter Notebook.




Q3. What is App routing in Flask? Why do we use app routes?
App Routing in Flask: App routing in Flask is a mechanism that maps URLs to specific functions in the application. It allows you to define different routes (URLs) in your application and associate them with functions that generate responses for those routes.

Why We Use App Routes:

URL Mapping: Routes define how URLs are mapped to functions, allowing different parts of the application to respond to different URLs.
Organization: Routes help in organizing the application's logic by separating concerns and allowing different functions to handle different URLs.
Dynamic Responses: Routes can include dynamic parts (e.g., URL parameters) that enable applications to generate responses based on user input or other variables.
User Interaction: Routes handle user requests and provide appropriate responses, facilitating interaction between users and the application.



Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/” route to show the following details:
Company Name: ABC Corporation
Location: India
Contact Detail: 999-999-9999
Code for Flask Application:

python
Copy code
from flask import Flask

app = Flask(__name__)

@app.route('/')
def home():
    return '''
    Company Name: ABC Corporation<br>
    Location: India<br>
    Contact Detail: 999-999-9999
    '''

@app.route('/welcome')
def welcome():
    return 'Welcome to ABC Corporation'

if __name__ == '__main__':
    app.run(debug=True)




    
Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.
Function for URL Building: The function used in Flask for URL building is url_for(). It generates a URL for a specific function based on the endpoint name.

Example Code Using url_for():


from flask import Flask, url_for, redirect

app = Flask(__name__)

@app.route('/')
def home():
    return 'Home Page'

@app.route('/welcome')
def welcome():
    return 'Welcome to ABC Corporation'

@app.route('/redirect-to-welcome')
def redirect_to_welcome():
    return redirect(url_for('welcome'))

if __name__ == '__main__':
    app.run(debug=True)
Explanation:

The url_for('welcome') function generates the URL for the welcome route.
The redirect(url_for('welcome')) function redirects the user to the /welcome route.
To test the url_for() function:

```
